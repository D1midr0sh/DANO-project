```python
data1 = data[data['button_tap_flg'] == 0]
data2 = data1[((data1['like_tap_flg'] == 1) | (data1['share_tap_flg'] == 1) | (data1['favorite_tap_flg'] == 1))]
print(data2.shape[0] / data1.shape[0])
```

```python
#индекс полезности для рекламы
w1 = data[(data.like_tap_flg == 1) & (data.share_tap_flg == 1)].shape[0] / data[data.like_tap_flg == 1].shape[0]
w2 = 1
w3 = data[(data.favorite_tap_flg == 1) & (data.share_tap_flg == 1)].shape[0] / data[data.favorite_tap_flg == 1].shape[0]
data['act_index'] = data.like_tap_flg * w1 + data.share_tap_flg * w2 + data.favorite_tap_flg * w3
weights = pd.Series({'Первый вес': round(w1, 4), 'Второй вес': w2, 'Третий вес': round(w3, 4)})
weights
```

```python
hours = data.groupby('user_hour', as_index=False).agg({'act_index': 'mean'})
for i in range(hours.shape[0]): 
    line = hours.iloc[i, 0]
    line = line[:len(line) - 2]
    hours.iloc[i, 0] = int(line) 
hours.sort_values('user_hour')
sns.scatterplot(data=hours, x='user_hour', y='act_index').set(title='Зависимость индекса от времени суток', xlabel='Час', ylabel='Среднее значение индекса')
```

```python
corr = np.corrcoef(list(hours['user_hour']), list(hours['act_index']))
print('Корреляция между часом и средним индексом:', round(corr[0][1], 2))
```
